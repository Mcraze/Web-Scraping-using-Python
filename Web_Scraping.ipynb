```python
#Importing Libraries
from bs4 import BeautifulSoup as bs
import requests
import pandas as pd
```

```python
#Setting up Url for Web Scraping
url = 'https://en.wikipedia.org/wiki/List_of_largest_companies_in_the_United_States_by_revenue'
```

```python
#Creating Page request
page = requests.get(url)
print(page)

#Getting data from the url as text
data = bs(page.text, 'html')
```

```python
#Checking data
print(data)
```

```python
#Making data more readable
increased_readability = data.prettify()
print(increased_readability)
```

```python
#Getting Table from the webpage

#Index is used in order to get the second table from the html page
table = data.find_all('table')[1]

print(table)
```

```python
#Getting headings from the selected table
titles = table.find_all('th')
print(titles)
```

```python
#Cleaning the headers data
table_titles = [x.text.strip() for x in titles]
print(table_titles)
```

```python
#Creating a dataframe using pandas and storing the headers
df = pd.DataFrame(columns = table_titles)

#Checking data
df
```

```python
#Scraping the rows of the data table using for loop and inserting it into dataframe
rows = table.find_all('tr')

#For loop needs to start from index 1 as the first row is header and contains th instead of td
for row in rows[1:]:
  row_data = row.find_all('td')
  each_row_data = [data.text.strip() for data in row_data]
  #Inserting data into the dataframe by checking length of data frame
  length = len(df)
  df.loc[length] = each_row_data
```

```python
#Checking the data in the data frame
df
```
